```python
#Convertir variables categ√≥ricas a variables dummy
data = pd.get_dummies(data, columns=['chest_pain_type', 'thallium'], drop_first=True)
```

```python
#Separar variables independientes (X) y dependiente (y)
X = data.drop(columns=['heart_disease', 'hd_dummy'])  # Eliminar la variable objetivo
y = data['hd_dummy']
```

```python
#Escalar las variables num√©ricas 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Dividir los datos en entrenamiento y prueba Divide los datos escalados:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
#Entrenar un modelo de clasificaci√≥n (Regresion logistica)
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

```

```python
#Evaluar el modelo Predicciones, matriz de confusi√≥n y m√©tricas:
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)

# M√©tricas
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
#Curva ROC y AUC Si quieres visualizar la Curva ROC:
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

y_proba = model.predict_proba(X_test)[:, 1]  # Probabilidad de la clase positiva
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
auc_score = roc_auc_score(y_test, y_proba)

# Gr√°fica
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'AUC = {auc_score:.2f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
